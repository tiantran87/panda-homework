```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
from pathlib import Path

%matplotlib inline
```

# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp_tsx_history.csv`: Contains historical closing prices of the S&P TSX 60 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data.

```python
# Reading whale returns
whale_returns_path = Path("./Resources/whale_returns.csv")
whale_data = pd.read_csv(whale_returns_path)
# Set the Date column to datetime format
whale_data["Date"] = pd.to_datetime(whale_data["Date"])
# Set whale_data index as Date
whale_data = whale_data.set_index("Date")
whale_data.head(10)
#whale_data.dtypes
```

```python
# Count nulls
whale_data.isnull().sum()
```

```python
# Drop nulls
whale_data = whale_data.dropna().copy()
whale_data
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data.

```python
# Reading algorithmic returns
algo_returns_path = Path("./Resources/algo_returns.csv")
algo_data = pd.read_csv(algo_returns_path)
# Set the Date column to datetime format
algo_data["Date"] = pd.to_datetime(algo_data["Date"])
# Set algo_data index as Date
algo_data = algo_data.set_index("Date")
algo_data.head(10)
```

```python
# Count nulls
algo_data.isnull().sum()
```

```python
# Drop nulls
algo_data = algo_data.dropna().copy()
algo_data
```

## S&P TSX 60 Returns

Read the S&P TSX 60 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P TSX 60 Closing Prices
sp_tsx_history_path = Path("./Resources/sp_tsx_history.csv")
sp_tsx_data = pd.read_csv(sp_tsx_history_path)
# Set the Date column to datetime format
sp_tsx_data["Date"] = pd.to_datetime(sp_tsx_data["Date"])
# Set sp_tsx_data index as Date
sp_tsx_data = sp_tsx_data.set_index("Date")
sp_tsx_data.head()
```

```python
# Check Data Types
sp_tsx_data.dtypes
```

```python
# Fix Data Types
sp_tsx_data["Close"] = sp_tsx_data["Close"].str.replace("$", "").str.replace(",","")
sp_tsx_data["Close"] = sp_tsx_data["Close"].astype("float")
sp_tsx_data.head(10)

```

```python
# Calculate Daily Returns
sp_tsx_data["Close"] = sp_tsx_data["Close"].pct_change()
sp_tsx_data
```

```python
# Drop nulls
sp_tsx_data = sp_tsx_data.dropna().copy()
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp_tsx_data = sp_tsx_data.rename(columns={"Close": "S&P TSX"})
sp_tsx_data
```

## Combine Whale, Algorithmic, and S&P TSX 60 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.
combine_porforlios_cols = pd.concat([whale_data, algo_data, sp_tsx_data], axis="columns", join="inner")
combine_porforlios_cols
```

## Risk Analysis

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios.
4. Determine which portfolios are riskier than the S&P TSX 60.
5. Calculate the Annualized Standard Deviation.

### Create a box plot for each portfolio


```python
# Box plot to visually show risk
sns.set(rc={'figure.figsize':(22,10)})
sns.boxplot(data=combine_porforlios_cols)
```

### Calculate Standard Deviations

```python
# Calculate the daily standard deviations of all portfolios
daily_std = combine_porforlios_cols.std()
daily_std

```

### Determine which portfolios are riskier than the S&P TSX 60

```python
# Calculate  the daily standard deviation of S&P TSX 60
sp__tsx_daily_std = combine_porforlios_cols["S&P TSX"].std()
sp__tsx_daily_std
# Determine which portfolios are riskier than the S&P TSX 60
print(f"The below portfolios are riskier than S&P TXS 60 (daily standard deviation: {sp__tsx_daily_std}): \n")
for i in range(len(daily_std)):
    if sp__tsx_daily_std < daily_std[i]:
        print(f"{daily_std.index[i]} with daily standard deviation: {daily_std[i]}")
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
annualized_std = daily_std * np.sqrt(252)
annualized_std
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for all portfolios using a 21-day window.
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P TSX 60.
3. Choose one portfolio, then calculate and plot the 60-day rolling beta for it and the S&P TSX 60.

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window

rolling_standard_21 = combine_porforlios_cols.rolling(window=21).std()

# Plot the rolling standard deviation

rolling_standard_21.plot(figsize=(30, 20))
```

### Calculate and plot the correlation

```python
# Calculate the correlation
portfolios_correlation = combine_porforlios_cols.corr()
print(portfolios_correlation)

# Display de correlation matrix
sns.heatmap(portfolios_correlation, vmin=-1, vmax=1)
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 60 TSX

```python
# Calculate covariance of a single portfolio
soros_rolling_covariance = combine_porforlios_cols["SOROS FUND MANAGEMENT LLC"].rolling(window=60).cov(combine_porforlios_cols["S&P TSX"])
#paulson_covariance = combine_porforlios_cols["PAULSON & CO.INC. "].rolling(window=60).cov(combine_porforlios_cols["S&P TSX"])
#tiger_covariance = combine_porforlios_cols["TIGER GLOBAL MANAGEMENT LLC"].rolling(window=60).cov(combine_porforlios_cols["S&P TSX"])
#berkshire_covariance = combine_porforlios_cols["BERKSHIRE HATHAWAY INC"].rolling(window=60).cov(combine_porforlios_cols["S&P TSX"])
#algo_1_covariance = combine_porforlios_cols["Algo 1"].rolling(window=60).cov(combine_porforlios_cols["S&P TSX"])
#algo_2_covariance = combine_porforlios_cols["Algo 2"].rolling(window=60).cov(combine_porforlios_cols["S&P TSX"])

# Calculate variance of S&P TSX
sp_rolling_variance = combine_porforlios_cols["S&P TSX"].rolling(window=60).var()

# Computing beta
soros_rolling_beta = soros_rolling_covariance / sp_rolling_variance
#paulson_beta = paulson_covariance / sp_variance
#tiger_beta = tiger_covariance / sp_variance
#berkshire_beta = berkshire_covariance / sp_variance
#algo_1_beta = algo_1_covariance / sp_variance
#algo_2_beta = algo_2_covariance / sp_variance

# Plot beta trend - calculate and plot the 60-day rolling beta for it and the S&P TSX 60.

soros_rolling_beta.plot(figsize=(20, 10), title="60-day rolling beta of Soros against S&P TSX 60")
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Use `ewm` to calculate the rolling window
combined_ewm = combine_porforlios_cols.ewm(halflife=21).mean()
combined_ewm.plot()
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratios = ((combine_porforlios_cols.mean()* 252) / (combine_porforlios_cols.std() * np.sqrt(252)))
sharpe_ratios

```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind="bar")
```

In this section, you will build your own portfolio of stocks, calculate the returns, and compare the results to the Whale Portfolios and the S&P TSX 60.

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.



```python
# Reading data from 1st stock
stock_1_path = Path("./Resources/TSLA.csv")
stock_1_data = pd.read_csv(stock_1_path)
# Set the Date column to datetime format
stock_1_data["Date"] = pd.to_datetime(stock_1_data["Date"]).dt.date
# Set index as Date
stock_1_data = stock_1_data.set_index("Date")
# Rename last column
stock_1_data = stock_1_data.rename(columns={"Close": "TSLA"})
stock_1_data
```

```python
# Reading data from 2nd stock
stock_2_path = Path("./Resources/AMZN.csv")
stock_2_data = pd.read_csv(stock_2_path)
# Set the Date column to datetime format
stock_2_data["Date"] = pd.to_datetime(stock_2_data["Date"]).dt.date
# Set index as Date
stock_2_data = stock_2_data.set_index("Date")
# Rename last column
stock_2_data = stock_2_data.rename(columns={"Close": "AMZN"})
stock_2_data.head(10)
```

```python
# Reading data from 3rd stock
stock_3_path = Path("./Resources/MSFT.csv")
stock_3_data = pd.read_csv(stock_3_path)
# Set the Date column to datetime format
stock_3_data["Date"] = pd.to_datetime(stock_3_data["Date"]).dt.date
# Set index as Date
stock_3_data = stock_3_data.set_index("Date")
# Rename last column
stock_3_data = stock_3_data.rename(columns={"Close": "MSFT"})
stock_3_data.head(10)
```

```python
# Combine all stocks in a single DataFrame
combine_stock_cols = pd.concat([stock_1_data, stock_2_data, stock_3_data], axis="columns", join="inner")
combine_stock_cols
```

```python
# Reset Date index
combine_stock_cols.sort_values("Date")
```

```python
# Calculate daily returns
combine_stock_daily_return = combine_stock_cols.pct_change()
# Drop NAs
combine_stock_daily_return = combine_stock_daily_return.dropna()
# Display sample data
combine_stock_daily_return.head(10)
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/3, 1/3, 1/3]
# Calculate portfolio return
weighted_returns = (weights * combine_stock_daily_return)
custom_port_return = weighted_returns.sum(axis=1)
custom_port_return = custom_port_return.to_frame(name="Custom Portfolio")
# Display sample data
custom_port_return.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
all_porforlios = pd.concat([combine_porforlios_cols, custom_port_return], axis="columns", join="inner")
all_porforlios
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
all_porforlios.dropna()
```

## Re-run the risk analysis with your portfolio to see how it compares to the others

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized `std`
all_porforlios_daily_std = all_porforlios.std()

all_porforlios_annual_std = all_porforlios_daily_std * np.sqrt(252)
all_porforlios_annual_std
```

### Calculate and plot rolling `std` with 21-day window

```python
# Calculate rolling standard deviation
all_port_rolling_standard_21 = all_porforlios.rolling(window=21).std()

# Plot rolling standard deviation
all_port_rolling_standard_21.plot(figsize=(30, 20))
```

### Calculate and plot the correlation

```python
# Calculate and plot the correlation
all_portfolios_correlation = all_porforlios.corr()
print(all_portfolios_correlation)
sns.heatmap(all_portfolios_correlation, vmin=-1, vmax=1)
```

### Calculate and Plot the 60-day Rolling Beta for Your Portfolio compared to the S&P 60 TSX

```python
# Calculate and plot Beta
# Calculate rolling covariance of custom portfolio
custom_rolling_covariance = all_porforlios["Custom Portfolio"].rolling(window=60).cov(all_porforlios["S&P TSX"])
# Calculate rolling variance of S&P TSX
sp60_rolling_variance = all_porforlios["S&P TSX"].rolling(window=60).var()
# Computing rolling beta
custom_rolling_beta = custom_rolling_covariance / sp60_rolling_variance
# Plot beta trend - calculate and plot the 60-day rolling beta for it and the S&P TSX 60.
custom_rolling_beta.plot(figsize=(20, 10), title="60-day rolling beta of Custom Stock against S&P TSX 60")
```

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Calculate Annualzied Sharpe Ratios
all_port_sharpe_ratios = ((all_porforlios.mean()* 252) / (all_porforlios.std() * np.sqrt(252)))
all_port_sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
all_port_sharpe_ratios.plot(kind="bar")
```
